# Correlation lengths plot

```python
import pickle
import matplotlib
import matplotlib.pyplot as plt
```

```python
with open(r'../exact_diagonalization/data/ed_correlation_lengths.pkl', 'rb') as f:
    ed_corr_length_data = pickle.load(f)
```

```python
with open(r'data/200_site_mps_correlation_lengths.pkl', 'rb') as f:
    mps_corr_length_data = pickle.load(f)
```

```python
fig, ax = plt.subplots(figsize=(6, 4))

ax.scatter(
    mps_corr_length_data['b_parameters'],
    mps_corr_length_data['correlation_lengths'],
    label=r'MPS',
    color='tab:blue'
)
ax.scatter(
    ed_corr_length_data['b_parameters'],
    ed_corr_length_data['correlation_lengths'],
    label=r'ED',
    color='tab:red'
)

ax.xaxis.set_major_locator(
    matplotlib.ticker.MultipleLocator(0.2)
)
ax.tick_params(
    axis='both',
    which='major',
    labelsize=11
)

ax.grid()

ax.legend(loc='lower right')

ax.set_xlabel(
    '$B$',
    fontsize=12
)
ax.set_ylabel(
    'Correlation length',
    fontsize=12
)

fig.savefig(r'plots/mps_ed_correlation_lengths.png', dpi=200)
```

```python
with open(r'data/ed_z_string_order_parameters.pkl', 'rb') as f:
    z_string_order_parameters_array = pickle.load(f)
```

```python
with open(r'data/ed_x_string_order_parameters.pkl', 'rb') as f:
    x_string_order_parameters_array = pickle.load(f)
```

```python
fig, ax = plt.subplots(figsize=(6, 4))

ax.scatter(
    ed_corr_length_data['b_parameters'],
    z_string_order_parameters_array,
    label=r'$\langle ZXIX\ldots IXZ \rangle $',
    color='tab:blue'
)
ax.scatter(
    ed_corr_length_data['b_parameters'],
    x_string_order_parameters_array,
    label=r'$\langle XXIX\ldots IXX \rangle $',
    color='tab:red'
)

ax.xaxis.set_major_locator(
    matplotlib.ticker.MultipleLocator(0.2)
)
ax.tick_params(
    axis='both',
    which='major',
    labelsize=11
)

ax.grid()

ax.legend(loc='upper center')

ax.set_xlabel(
    '$B$',
    fontsize=12
)
ax.set_ylabel(
    'String order parameter',
    fontsize=12
)

#fig.savefig(r'plots/ed_Z_X_string_order_parameter.png', dpi=200)
```

```python
fig, (ax1, ax2) = plt.subplots(figsize=(6, 6), nrows=2)

ax1.plot(
    mps_corr_length_data['b_parameters'],
    mps_corr_length_data['correlation_lengths'],
    label=r'MPS',
    color='tab:blue',
    marker='D'
)
ax1.plot(
    ed_corr_length_data['b_parameters'],
    ed_corr_length_data['correlation_lengths'],
    label=r'ED',
    color='tab:red',
    marker='D'
)

ax1.xaxis.set_major_locator(
    matplotlib.ticker.MultipleLocator(0.2)
)
ax1.xaxis.set_minor_locator(
    matplotlib.ticker.MultipleLocator(0.1)
)
ax1.tick_params(
    axis='y',
    which='major',
    labelsize=11
)

ax1.grid()

ax1.legend(
    loc='lower center',
    fontsize=12
)

ax1.set_ylabel(
    'Correlation length',
    fontsize=12
)

ax1.set_xticklabels([])

ax2.plot(
    ed_corr_length_data['b_parameters'],
    z_string_order_parameters_array,
    label=r'$\langle ZXIX\ldots IXZ \rangle $',
    color='tab:blue',
    marker='D'
)
ax2.plot(
    ed_corr_length_data['b_parameters'],
    x_string_order_parameters_array,
    label=r'$\langle XXIX\ldots IXX \rangle $',
    color='tab:red',
    marker='D'
)

ax2.xaxis.set_major_locator(
    matplotlib.ticker.MultipleLocator(0.2)
)
ax2.xaxis.set_minor_locator(
    matplotlib.ticker.MultipleLocator(0.1)
)
ax2.tick_params(
    axis='both',
    which='major',
    labelsize=11
)

ax2.grid()

ax2.legend(
    loc='center left',
    fontsize=12
)

ax2.set_xlabel(
    '$B$',
    fontsize=12
)
ax2.set_ylabel(
    'String order parameter',
    fontsize=12
)

fig.subplots_adjust(hspace=0.1)


fig.savefig(r'plots/ed_Z_X_string_order_parameter_and_correlation_lengths.png', dpi=200)
```
