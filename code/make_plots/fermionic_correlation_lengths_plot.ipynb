# Fermionic correlation lengths plot

```python
import pickle

import numpy as np

import matplotlib
import matplotlib.pyplot as plt

from matplotlib.ticker import MultipleLocator
```

```python
with open(r'data/fermionic_cases/200_site_mps_iz_triv_correlation_lengths.pkl', 'rb') as f:
    triv_corr_length_data = pickle.load(f)
```

```python
with open(r'data/fermionic_cases/200_site_mps_zz_non_triv_correlation_lengths.pkl', 'rb') as f:
    non_triv_corr_length_data = pickle.load(f)
```

```python
non_triv_corr_length_data
```

```python
interpolation_values = np.linspace(0, 1, 21)
interpolation_values = np.delete(interpolation_values, 10)

sub_triv_corr_lengths = np.delete(triv_corr_length_data['correlation_lengths'], 10)[::-1]
sub_non_triv_corr_lengths = np.delete(non_triv_corr_length_data['correlation_lengths'], 10)[::-1]
```

```python
fig, ax = plt.subplots(figsize=(6, 4))

ax.plot(
    interpolation_values,
    sub_triv_corr_lengths,
    label=r'$H_{f,1}(t), IZ$',
    color='tab:blue',
    marker='D'
)
ax.plot(
    interpolation_values,
    sub_non_triv_corr_lengths,
    label=r'$H_{f,2}(t), ZZ$',
    color='tab:red',
    marker='D'
)

ax.grid()

ax.set_xticks(interpolation_values)

ax.xaxis.set_major_locator(MultipleLocator(0.1))
ax.xaxis.set_minor_locator(MultipleLocator(0.05))
ax.tick_params(
    axis='both',
    which='major',
    labelsize=14
)


ax.yaxis.set_major_locator(MultipleLocator(2))
ax.yaxis.set_minor_locator(MultipleLocator(1))

ax.set_xlabel("t", fontsize=15)
ax.set_ylabel("Correlation Length", fontsize=15)

ax.legend(
    fontsize='x-large'
)

fig.savefig(
    'plots/fermionic_state_correlation_lengths.png',    
    dpi=200,
    bbox_inches='tight'
)
```
